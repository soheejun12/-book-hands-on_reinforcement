# Q학습을 사용하여 taxi 문제 해결하기 
#### - 에이전트= 택시 운전사 
#### - 4가지 위치가 존재 
#### - 에이전트는 한 곳에서 승객을 태우고 다른 곳에 내려줘야 함 
#### - 올바른 하차에는 보상 +20점
#### - 승객을 내려주러 가는 동안 걸리는 각 time step은 -1점 
#### - 잘못된 (불법) 승차와 하차는 -10점 
#### - 목적은 불법 승객없이 최단시간으로 승객을 올바른 곳에서 승차하고 하차하는 것 

```python
import gym
import random 
```

```python
#env = gym.make("Taxi-v1") #오류  
env = gym.make("Taxi-v2")
```

```python
env.render()
```

- R, G, Y, B: 4가지 위치
- 노란 사각형: 택시 (에이전트)

```python
alpha = 0.4
gamma = 0.999
epsilon = 0.017
```

- Q 테이블 초기화 
    - {(state, action): Q 값} 형태의 사전 구조 

```python
q = {}

for s in range(env.observation_space.n):
    for a in range(env.action_space.n):
        q[(s, a)] = 0.0
```

- Q학습 업데이트 규칙을 통해 Q테이블을 업데이트하는 함수 정의 
<img src = "./image/Q.PNG">

    - qa: 다음 상태에서 가능한 액션들에 대한 Q값중 최대값 

```python
def update_q_table(prev_state, action, reward, nextstate, alpha, gamma):
    qa = max([q[(nextstate, a)] for a in range (env.action_space.n)])
    
    q[(prev_state, action)] += alpha * (reward + gamma * qa - q[(prev_state, action)])
```

- epsilon greedy 정책을 수행하는 함수 정의 
    - 랜덤값을 생성하여 그 값이 epsilon보다 작으면 탐험을, 크면 탐색을 진행 

```python
def epsilon_greedy_policy(state, epsilon):
    if random.uniform(0, 1) < epsilon: #0~1 균일분포 
        return env.action_space.sample() #탐험 
    
    else: 
        return max(list(range(env.action_space.n)), 
                   key=lambda x: q[(state, x)])
    
```

- Q학습

```python
for i in range(8000):
    #초기화
    r = 0
    prev_state = env.reset()
    
    while True:
        #epsilon greedy로 액션 선택 
        action = epsilon_greedy_policy(prev_state, epsilon)
        
        #액션 수행 
        nextstate, reward, done, _ = env.step(action)
        
        #Q값 업데이트
        update_q_table(prev_state, action, reward, nextstate, alpha, gamma)
        
        #S'가 S가 됨 
        prev_state = nextstate
        
        r += reward
        
        if done: 
            break 
            
        print((i+1), "total reward: ", r)
        
env.close()
```
