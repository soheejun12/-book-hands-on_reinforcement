# DARQN 
### : Deep Attention Recurrent Q Network 

### DQRN의 확장 
#### - 가장 위 convolutional layer에 attention layer 추가 
- attention mechanism 
     - 이미지 캡셔닝, 물체 탐지에 사용 
     - 신경망에서 이미지에 캡션을 달기 위해서는 이미지에 있는 세부적인 특정 물체에 집중해야함 
     - 세부적인(specific) 물체에 집중하도록 해주는 역할 
     
     
#### - DARQN = DRQN + Attention layer 
- 이미지의 작은 구역(region)에 집중 가능 
    - 네트워크 파라미터 수 감소 
    - 학습, 테스트 시간 감소 
    
  
- DARQN의 LSTM layer
    - 다음 최적 액션을 위해 이전 상태 정보 저장 
    - (추가) 다음 집중할 이미지 구역을 결정하는 정보 저장

###  3개 층 
#### - convolution layer 
- 게임 화면을 입력으로 받음  
- 특징맵 생성 
    

#### -  attention layer
- 특징맵을 context vector로 변환
- context vector : linear combination의 결과 
   - 이전 은닉상태를 포함하는 벡터 
        
     
#### - LSTM layer 
- context vector를 입력으로 받음 
- 2개의 출력 
   - 현재 상태에서, 어떤 액션을 취할지 결정하는 Q값 
   - 다음 time step에서, attention 네트워크가 이미지의 어떤 구역에 집중할지 결정하는데 도움을 주는 값 
      - 더 나은 context vector 생성 가능 
      
      
## * attention 종류
#### - Soft attention
- CNN에 의해 생성된 모든 특징맵의 가중화된 평균값이 context vector 
    - 가중화는 특징의 연관 중요도(relative importance)에 따라 선택됨 
    

#### - Hard attention
- 구역 선택(location selection)정책 파이에 따라, time step t에서 이미지의 특정 구역에 집중 
    - 정책 파이는, 신경망으로 표현
        - 가중치: 정책 파라미터 
        - 출력: 구역(location) 선택 확률 
        
- 하지만, soft attention보다 그렇게 좋은 성능을 보이진 않음
