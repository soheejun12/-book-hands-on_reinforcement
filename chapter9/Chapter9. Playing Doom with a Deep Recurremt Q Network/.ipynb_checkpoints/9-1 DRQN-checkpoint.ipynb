# DRQN (Deep Recurrent Q Network)

- DQN이 인간의 수준에서 아타리게임을 할 수 있는데, DRQN이 필요한 이유? 
    - POMDP 문제 
    
## * POMDP (Partially Observable Markov Decision Process)
#### : 환경으로부터 제한된 정보만을 얻을 수 있는 경우 
#### - 지금까지 살펴 본 경우는 전체 관찰 가능한 MDP 
- 모든 가능한 액션과 상태에 대한 정보를 알고있음 
- 비록 에이전트는 전이확률과 보상확률에 대해 알지 못한다고 해도 환경에 대한 완전한 정보를 알고 있음 
    - frozenlake환경 
        - 모든 상태(에이전트의 위치)와 액션(이동방향)에 대해 알 수 있음 
        
        
- 쉽게 환경을 모델링할 수 있음 
- 하지만 거의 대부분의 실제환경은 부분적으로 관찰 가능 
    - 모든 상태를 볼 수 없음 
    - 실제환경에서 에이전트가 걷는 것을 배울 때, 에이전트는 환경에 대한 모든 정보를 알 수 없음 
    - 시야 밖에 대한 정보를 모름 
    
#### - 부분적으로 관찰 가능한 POMDP 
- 상태는 부분적인 정보만을 제공 
- 과거 상태에 대한 정보를 메모리에 저장해 둠 
    - 환경의 본질에 대한 이해를 돕기위해 
    - 정책을 발전시키기 위해  
    
- 그러므로, POMDP환경에서는 최적의 액션을 수행하기 위해서 이전 상태에 대한 정보를 재학습해야 함 

#### -  예시) pong game
- 현재화면만 고려한다면, 
    - 공의 위치만 파악 가능 
    - 움직이는 공의 방향, 속도 파악 불가능 
 
- 그러므로, 과거 4개의 화면을 고려 
    - 공의 방향, 속도 파악 가능 
    
    
- DQN 
    - 현재화면과 함께, 과거 4개의 게임 화면을 convolution 레이어의 입력으로 전달 
    
  
- 하지만, 과거 4개의 화면을 사용하는 것만으로 다른 환경을 이해하는 것에 도움이 될까?
    - 어떤 환경은 과거 100개의 화면을 필요로할 수도 있음 
    - 하지만, n개의 게임 화면을 쌓는 것은 학습 과정을 느리게하고 경험재생버퍼의 크기를 증가시킴 
    
    
- 그러므로, RNN의 장점을 통해 이전상태를 필요로 할 때까지 이해하고 유지하자 

- 7장에서 배웠던 LSTM(Long Short-Term Memory)
    - 텍스트 생성
    - 필요할 때까지 정보의 유지, 망각, 갱신을 통해 단어의 문맥을 이해 
    

- 과거의 정보를 이해하기 위해 DQN구조를 수정
    - convolitional 후(post) 첫번째 fully connected 레이어를 LSTM RNN레이어로 대체 
    - 이것을 통해 에이전트가 과거에 대한 상태를 기억할 수 있어 POMDP 문제를 해결할 수 있고 정책을 향상시킬 수 있음 

# DRQN 구조 

#### DQN과 비슷하지만, 첫번째 fc 레이어가 LSTM RNN으로 대체됨 

<img src="./img/drqn.png">

- 게임 화면을 convolutional 레이어의 입력으로 전달
    - convolution 레이어가 이미지를 합성곱하고 특징맵을 생성


- 특징맵의 결과가 LSTM 레이어로 전달 됨 
    - LSTM 레이어는 정보를 저장해놓는 메모리를 가지고 있음 
    - LSTM 레이어가 중요한 과거 게임 상태에 대한 정보를 유지하고 시간이 지남에때라 메모리를 갱신함 
    
    
- fully connected 레이어를 거치면 Q값이 출력됨 


- 그러므로, DQN과 달리 직접 Q($s_t$, $a_t$)를 추정하지 않음 
    - 대신, Q($h_t$, $a_t$)를 추정함 
        - $h_t$는 이전 time ste
        - $h_t$ = LSTM($h_t$$_-$$_1$, $o_t$)//
        
        
- 역전파를 통해 네트워크 학습 

#### 경험 재생 버퍼 (experience replay buffer)
- DQN에서는 
    - 연관성 있는 경험을 피하기 위해 게임 전이(transition)을 저장해놓은 경험 재생을 사용
    - 경험의 랜덤 batch를 사용해 네트워크를 학습 
    
- DRQN은, 
    - 경험 버퍼에 전체 에피소드를 저장 
    - 에피소드의 랜덤 batch로 부터 n개의 스텝을 랜덤하게 샘플링 
    - 이것을 통해 임의성(randomization)과 연속적인 경험을 모두 수용(accomodate) 가능

### Basic Doom Game 
- 먼저 간단한 예제를 통해 vizdoom 환경에 익숙해져보자 
- http://www.cs.put.poznan.pl/visualdoomai/tutorial.html
    - button : (action) 에이전트에 의해 눌러지는 게버튼 (에이전트가 수행하는 액션)

```python
from vizdoom import *
import random 
import time

game = DoomGame() #DoomGame 객체 생성

#vizdoom에서 제공하는 시나리오 중 basic 시나리오 로드
#시나리오 파일이 저장된 폴더 경로 입력 
game.load_config("/home/sohee/openai/ViZDoom/scenarios/basic.cfg") 

#시나리오를 통해 게임 초기화 
game.init()
```

```python
#one-hot encoding된 액션 
shoot = [0, 0, 1]
left = [1, 0, 0]
right = [0, 1, 0]

actions = [shoot, left, right]
```

```python
actions
```

```python
no_of_episodes = 10

for i in range(no_of_episodes):
    
    #새로운 에피소드 시작, 
    game.new_episode()
    
    while not game.is_episode_finished():
        
        #get_state(): GameState 객체 반환 
        state = game.get_state()
        
        img = state.screen_buffer
        
        misc = state.game_variables
        
        #액션 수행 
        reward = game.make_action(random.choice(actions))
        
        print(reward)
        
    #2초동안 실행 일시중지     
    time.sleep(2)
```
